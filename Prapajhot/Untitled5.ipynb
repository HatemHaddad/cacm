```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# Download historical stock prices from Yahoo Finance
ticker = "AAPL" # Change to desired stock ticker
start_date = "2020-01-01" # Change to desired start date
end_date = "2022-02-23" # Change to desired end date
stock_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
stock_data
```

```python
# Calculate 20-day and 50-day moving averages
stock_data['MA20'] = stock_data['Close'].rolling(window=20).mean()
stock_data['MA50'] = stock_data['Close'].rolling(window=50).mean()

# Define function to execute trade based on cross-over
```

```python
stock_data
```

```python
def execute_trade(row):
    if row['MA20'] > row['MA50']:
        # Buy if 20-day moving average crosses above 50-day moving average
        row['Position'] = 1
        row['Entry Price'] = row['Close']
    elif row['MA20'] < row['MA50']:
        # Sell if 20-day moving average crosses below 50-day moving average
        row['Position'] = -1
        row['Entry Price'] = row['Close']
    return row

# Apply function to each row of stock data
stock_data = stock_data.apply(execute_trade, axis=1)
```

```python
stock_data
```

```python
# Initialize variables for tracking trades
position = 0
entry_price = 0
stop_loss = 0.05 # Change to desired stop-loss percentage
take_profit = 0.1 # Change to desired take-profit percentage
```

```python
# Define function to calculate trade outcomes
def calculate_outcome(row):
    global position, entry_price
    if row['Position'] != position:
        # Close existing position if cross-over occurs
        if position == 1:
            # Sell if position is long
            row['Exit Price'] = row['Close']
            row['Outcome'] = (row['Exit Price'] - entry_price) / entry_price
            if row['Outcome'] < -stop_loss:
                # Set stop loss order if trade is losing
                row['Exit Price'] = entry_price * (1 - stop_loss)
                row['Outcome'] = -stop_loss
            elif row['Outcome'] > take_profit:
                # Set take-profit order if trade is winning
                row['Exit Price'] = entry_price * (1 + take_profit)
        elif position == -1:
            # Buy if position is short
            row['Exit Price'] = row['Close']
            row['Outcome'] = (entry_price - row['Exit Price']) / entry_price
            if row['Outcome'] < -stop_loss:
                # Set stop loss order if trade is losing
                row['Exit Price'] = entry_price * (1 + stop_loss)
                row['Outcome'] = -stop_loss
            elif row['Outcome'] > take_profit:
                # Set take-profit order if trade is winning
                row['Exit Price'] = entry_price * (1 - take_profit)
        position = row['Position']
        entry_price = row['Entry Price']
    return row
```

```python
# Apply function to each row of stock data
stock_data = stock_data.apply(calculate_outcome, axis=1)
```

```python
stock_data
```

```python
# Calculate cumulative returns
stock_data['Cumulative Returns'] = (1 + stock_data['Outcome']).cumprod()
```

```python
# Print final portfolio statistics
print("Total trades:", stock_data['Outcome'].count())
print("Winning trades:", (stock_data['Outcome'] > 0).sum())
print
```
