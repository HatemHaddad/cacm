```python
payroll_2021 = payroll[payroll['PayYear'] == '2021']
```

```python
payroll_2021 = payroll_2021[['EmployeeNumber','PayMonthNum','Gross']]
```

```python
payroll_2021 = payroll_2021.pivot_table('Gross',index='EmployeeNumber',columns='PayMonthNum').reset_index()
```

```python
payroll_2021
```

```python
merge_emp_payroll = employee_copy.merge(payroll_2021,left_on='EmpNo',right_on='EmployeeNumber',how='outer')
```

```python
ap = ap.astype({'Payment Date':'datetime64'})
ap['Year'] = ap['Payment Date'].dt.year
ap = ap[ap['Year'] == 2021]
```

```python
merged_payroll_ap = merge_emp_payroll.merge(ap,left_on='EmployeeName',right_on='Supplier Name',how='left')
merged_payroll_ap.sort_values('Invoice Amount',ascending=False).to_excel('invoices_sorted_2021.xlsx')
```

```python
merged_payroll_ap.groupby('EmployeeName').agg({'Invoice Amount':'sum'}).sort_values('Invoice Amount',ascending=False).to_excel('invoice_amount_1.xlsx')
```
