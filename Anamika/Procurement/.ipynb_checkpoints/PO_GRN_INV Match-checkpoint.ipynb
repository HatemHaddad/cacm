```python
# Make a copy from the AP and PO tables by using the copy() method.

ap_c = ap.copy()
po_c = po.copy()
```

```python
# PO - Date filtered after 2020-01-01
po_c = po_c[po_c['CREATION_DATE'] >= '2020-01-01']
```

```python
# AP - Date filtered after 2020-01-01
ap_c = ap_c[ap_c['CREATION_DATE'] >= '2020-01-01']
```

```python
# Groupby PO_Number and SUM by 'INVOICE_DIST_BASE_AMOUNT' as this is the converted currency to AED.

df_ap = ap_c.groupby('PO_NUMBER').agg({'INVOICE_DIST_AMOUNT':'sum'}).reset_index()
```

```python
# Calculate po['QUANTITY_ORDERED'] * po['UNIT_PRICE'] in the PO table.
```

```python
po_c['PO_Total'] = po_c['QUANTITY_ORDERED'] * po_c['UNIT_PRICE']
```

```python
df_po = po_c.groupby('PO_NUMBER').agg({'PO_Total':'sum'}).reset_index()
```

```python
df_po_ap = df_po.merge(df_ap,on='PO_NUMBER')
```

```python
df_po_ap['Diff'] = df_po_ap['PO_Total'] - df_po_ap['INVOICE_DIST_AMOUNT']
```
