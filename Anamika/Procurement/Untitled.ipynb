```python
df = yf.download('tqqq',start='2018-01-01')
```

```python
df['cc_return'] = df['Adj Close'].pct_change()
df['cc_return'] =1 + df['cc_return']
```

```python
df['Equity'] = df['cc_return'].cumprod()*100
df['Equity'].iloc[0] = 100
```
