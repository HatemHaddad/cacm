```python
import pandas as pd
import numpy as np
```

```python
sheet1 = pd.read_excel('2021-2022 POs (Contracts).xlsx',sheet_name='PO Details')
```

```python
sheet1 = sheet1.drop_duplicates(subset='Purchase Order Number')
```

```python
sheet21 = pd.read_excel('Sourcing (competetive vs. single)_old.xlsx',sheet_name='2021')
```

```python
sheet21 = sheet21.drop_duplicates(subset='PO#')
```

```python
sheet22 = pd.read_excel('Sourcing (competetive vs. single)_old.xlsx',sheet_name='2022')
```

```python
sheet22 = sheet22.drop_duplicates(subset='PO#')
```

```python
sheet2 = pd.concat([sheet21,sheet22]).drop_duplicates(subset='PO#')
```

```python
sheet1
```

```python
sheet1['Purchase Order Number'] = sheet1['Purchase Order Number'].astype('Int64').astype(str)  
```

```python
sheet1.info()
```

```python
sheet1.merge(sheet2,left_on='Purchase Order Number',right_on='PO#',how='outer').to_excel('merge1.xlsx')
```

```python
m = pd.read_excel('merge1.xlsx',index_col=0)
```

```python
p = pd.read_excel('Procurement 2021-2022 POs (Variation  Type) 16 June 2022.xlsx')
```

```python
p = p.drop_duplicates(subset='Purchase Order Number')
```

```python
p['Purchase Order Number'] = p['Purchase Order Number'].astype('Int64').astype(str)  
```

```python
p.merge(m,left_on='Purchase Order Number',right_on='Purchase Order Number',how='outer').to_excel('final.xlsx')
```
